```python
import pandas as pd
```

```python
messages = pd.read_csv('/content/SMSSpamCollection.txt', sep='\t', names=['label', 'message'])
messages.head()
```

```python
messages.shape
```

```python
# Data Cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(len(messages)):
  review = re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus[:5]
```

```python
# Bag of words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, binary=True)
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
# label encoding for y
y = pd.get_dummies(messages['label'])
y = y.iloc[:,1].values
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0, test_size=0.2)
```

```python
# training the multibionomialNB
from sklearn.naive_bayes import MultinomialNB
model_nb = MultinomialNB()
model_nb.fit(X_train, y_train)
```

```python
y_pred = model_nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(accuracy_score(y_pred, y_test))
print(classification_report(y_test, y_pred))
```

```python
# creating tfidf model
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2500)
X = tv.fit_transform(corpus).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0, test_size=0.2)
```

```python
model_nb = MultinomialNB()
model_nb.fit(X_train, y_train)
```

```python
y_pred = model_nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(accuracy_score(y_pred, y_test))
print(classification_report(y_test, y_pred))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[:5]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
words = []
for sent in corpus:
  sent_token = sent_tokenize(sent)
  for wrd in sent_token:
    words.append(simple_preprocess(wrd))
```

```python
words[:10]
```

```python
from gensim.models import Word2Vec
model = Word2Vec(words, window=5, min_count=2)
```

```python
model.wv.index_to_key
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.wv.similar_by_word('happy')
```

```python
import numpy as np
def avg_word2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```

```python
X=[]
for i in range(len(words)):
    X.append(avg_word2vec(words[i]))
```
